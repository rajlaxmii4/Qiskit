```python
qc=QuantumCircuit(1,1)
v=[1/sqrt(2),1/sqrt(2)]
qc.initialize(v,0)
qc.draw()
```

```python
psi=Statevector(v)
print(psi)
```

```python
D=DensityMatrix(psi)
print(D)
```

```python
vn_entropy1=entropy(psi,base=2)
vn_entropy2=entropy(D,base=2)
print(vn_entropy1)
print(vn_entropy2)
```

```python
qc=QuantumCircuit(2,2)
qc.h(0)
qc.cx(0,1)
qc.draw()
```

```python
sim=Aer.get_backend('aer_simulator')
qc.save_statevector()
qobj=assemble(qc)
result=sim.run(qobj).result()
out_state=result.get_statevector()
print("Bell State 00=", str(out_state))
```
